```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\Parth Patel\Downloads\raw_data_final.csv")
data
```

```python
data.head(15)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
data.head(15)
```

```python
categorical_cols = ['area_name', 'lighting_condition', 'transport_mode']
for col in categorical_cols:
    data[col] = data[col].fillna('Unknown')
```

```python
data
```

```python
data['crowd_density'] = data['crowd_density'].fillna(data['crowd_density'].median())
data['report_delay_minutes'] = data['report_delay_minutes'].fillna(data['report_delay_minutes'].median())
```

```python
data
```

```python
data = data.drop_duplicates()
```

```python
data
```

```python
data.to_csv('cleaned_data.csv', index=False)
```
